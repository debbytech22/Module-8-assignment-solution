```python
import matplotlib
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
tips = sns.load_dataset('tips')

tips
```

```python
plt.figure(figsize=(8,6))
ax=sns.kdeplot(tips['size'])
ax.set_title('KDE plot for the size feature of the tips dataset')
plt.show()
```

The KDE plot above shows that the graph of size has 3 mode.
Size 2 has the highest frequency.
The graph is skewed to the left.
It has outliers at 5,6,7

```python
plt.figure(figsize=(8,6))
ax=sns.boxplot(data=tips, x='sex', y='total_bill')
ax.set_title('Distribution of total bill by gender')
plt.show()
```


From the graph of distribution of total bill by gender plotted above,we can deduce that:
The maximum bill incurred by female is 30,.
There are outliers around 35 and 44 for females
The minimum bill by female is around 2.
For female the 25th percentile is 14 i.e 25% of male incurred less than 14 total bill
For female the median value is 16
The 75th is 22 i.e 75% of female incurred less than 22 total bill
For male the 25th percentile is 12 i.e 25% of male incurred less than 12 total bill
For male the median value is 18
The 75th is 25 i.e 75%of male incurred less than 25 total bill
There are outliers up to 50 for males
The minimum bill incurred by male is around 8
The maximum bill by male is a little above 40
This implies that males incurred more bill than females.


```python
plt.figure(figsize=(8,6))
ax=sns.boxplot(data=tips,x='day',y='tip')
ax.set_title('Distribution of tips by days')
plt.show()
```

From the graph of distribution of tips by days plotted above,we can deduce that:
1.No tip was given on monday,tuesday and wednesday.
2.The highest number of tip was given on sunday 
3.There are outliers on saturday that gave up to 10 tip 
4.All the days have almost the same min number of tips collected
5.The median values for all the days are 2.2, 3, 2.8, 3.2 respectively
6.The 75th percentile for thursday,friday,saturday is the same and the value is 3.6,for sunday is 4
6.The 25th percentile for all the days is 2.Therefore more tip was given on saturday due to the outlier


```python
plt.figure(figsize=(8,6))
ax=sns.boxplot(data=tips,x='day',y='total_bill',hue='smoker')
ax.set_title('Distribution of total bill by days considering smoker status of the customer')
plt.show()
```

From the graph of distribution of total bill by days considering smoker status of the customer plotted above,we can deduce that:
1.On thurday,the least bill incurred by smoker is 10 and the least bill incurred by non smokers is 7.The maximum bill incurred by the smokers was 28 while that of smokers is 30.Smokers and non smokers have almost the same median and 75th percentile.
2.On Friday,The least bill incurred by smokers is 5 while the least bill incurred by by non smokers is around 14.The median value of smoker is 14 while the median value of smoker is non smoker is 20,The max bill incurred by smokers is 28 while that of non smokers is 24.
3.On Saturday,the min incurred by smokers is 4 while non smokers is 8,The median value of smokers and 75th percntile of non smokers is the same,the highest bill incurred by smokers is around 44 with ouliers up to 52 while the highest bill by non smokers is 28 with outlier up to 48.
4.On sunday,the min bill by smokers is 6 while non smokers is 8,the median value for smokers is 22 while non smokers is 26,the 75th percentile for smokers is 30 while that of non smokers is 22,the highest bill incurred by smokers is around 45 while that of non smokers is around 37 with outlier.
5.I can deduce that smokers spend more on saturday and sunday
