```python
#Import Modules for project ( Matplotlib, Seaborn and Pandas)
import matplotlib
from matplotlib import pyplot as plt
import pandas as pd 
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('WorldCupMatches.csv')
df.head()
#df['Datetime'].dtype
```

```python
df.isnull().any()
```

```python
df['Year'].fillna(0,inplace=True)
df['Datetime'].fillna(0,inplace=True)
df['Stage'].fillna('None',inplace=True)
df['City'].fillna('None',inplace=True)

```

```python
df['Total Goals']= df['Home Team Goals'] + df['Away Team Goals']
df['Total Goals']
```

```python

df.head()
```

```python
ax=sns.barplot(data=df,x='year',y='Total Goals',ci= None)
sns.set_style('whitegrid')
ax.set_title('Tota Goals Scored each year between 1930-2014')
ax.set_xticklabels(ax.get_xticklabels(),rotation = 90)
plt.show()
```

```python
plt.figure(figsize=(8,6))
ax=sns.boxplot(data=df, x='year', y='Total Goals')
ax.set_title('Distribution of average goals for each year')
ax.set_xticklabels(ax.get_xticklabels(),rotation = 90)
plt.show()
```
